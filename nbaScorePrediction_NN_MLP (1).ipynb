```python
# Extract features (X) and target (y)
X = data.iloc[:, 0:17]  # Features from index 0 to 16
y = data.iloc[:, 17]    # Target (index 17)

# Optionally include additional features (index 18 to 31)
X_additional = data.iloc[:, 18:32]  # Additional features
```

```python
X.shape
```

```python
X_additional.shape
```

```python
X = pd.concat([X, X_additional], axis=1)
```

```python
X.shape
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error


# Apply standardization or normalization to features (if necessary)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # Standardization

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Create and train the MLP model
mlp = MLPRegressor(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=500, random_state=42)
mlp.fit(X_train, y_train)

# Make predictions
y_pred = mlp.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
import pandas as pd

# Load the new data from the CSV file into a DataFrame, skipping the first row
new_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/csv_nop.csv', skiprows=1)

# Apply the same preprocessing steps as done during training

# Standardize the new data using the same scaler used during training
X_new_scaled = scaler.transform(new_data.iloc[1:4])  # Assuming 'scaler' is the StandardScaler object used during training

# Iterate over each row in the standardized new data
for i, row in enumerate(X_new_scaled):  # Iterate over the first 3 rows
    # Use the trained MLP model to make predictions for the current row
    prediction = mlp.predict(row.reshape(1, -1))  # Reshape to 2D array for single row prediction
    # Print the prediction for the current row
    print(f"Prediction for row {i + 1}: {prediction[0]}")
```
